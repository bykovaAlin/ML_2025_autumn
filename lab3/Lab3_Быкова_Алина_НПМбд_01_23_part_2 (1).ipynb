## 2. Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. В этом разделе мы этим и займёмся.

У matplotlib, конечно, же есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторое величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить каждый из них. Не менее важно усвоить базовые принципы визуализаций:
- на графиках должны быть подписаны оси;
- у визуализации должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах).
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать.

#### 3. [1 балл] Постройте средних сумм покупок по дням недели (bar plot).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение файла с данными
df = pd.read_csv('chipotle.tsv', sep='\t', on_bad_lines='skip')

# Очистка столбца 'item_price' от символов '$', пробелов и запятых
df['item_price'] = df['item_price'].astype(str)
df['item_price'] = df['item_price'].str.replace('$', '', regex=False)
df['item_price'] = df['item_price'].str.replace(',', '', regex=False)
df['item_price'] = df['item_price'].str.strip()
df['item_price'] = pd.to_numeric(df['item_price'], errors='coerce')

# Преобразование столбца 'date' в формат даты
df['date'] = pd.to_datetime(df['date'])

# Добавляем столбец с днем недели
df['day_of_week'] = df['date'].dt.day_name()

# Создаем правильный порядок дней недели
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Рассчитываем среднюю сумму покупки по дням недели
avg_sales_by_day = df.groupby('day_of_week')['item_price'].mean()

# Упорядочиваем по правильному порядку дней недели
avg_sales_by_day = avg_sales_by_day.reindex(days_order)

# Построение bar plot
plt.figure(figsize=(12, 7))
bars = plt.bar(avg_sales_by_day.index, avg_sales_by_day.values,
               color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2'],
               edgecolor='black', alpha=0.8)

# Добавляем значения на столбцы
for bar, value in zip(bars, avg_sales_by_day.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
             f'${value:.2f}', ha='center', va='bottom', fontweight='bold')

# Настройки графика
plt.title('Средние суммы покупок по дням недели', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('День недели', fontsize=12)
plt.ylabel('Средняя сумма покупки ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

print("=" * 50)
print("СРЕДНИЕ СУММЫ ПОКУПОК ПО ДНЯМ НЕДЕЛИ")
print("=" * 50)
for day, avg_price in avg_sales_by_day.items():
    print(f"{day:12}: ${avg_price:.2f}")

print(f"Самый прибыльный день: {avg_sales_by_day.idxmax()} (${avg_sales_by_day.max():.2f})")
print(f"Наименее прибыльный день: {avg_sales_by_day.idxmin()} (${avg_sales_by_day.min():.2f})")
print(f"Разница: ${avg_sales_by_day.max() - avg_sales_by_day.min():.2f}")
```

#### 4. [1 балл] Постройте график зависимости денег за товар от купленного количества (scatter plot).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Чтение файла с данными
df = pd.read_csv('chipotle.tsv', sep='\t', on_bad_lines='skip')

# Очистка столбца 'item_price'
df['item_price'] = df['item_price'].astype(str)
df['item_price'] = df['item_price'].str.replace('$', '', regex=False)
df['item_price'] = df['item_price'].str.replace(',', '', regex=False)
df['item_price'] = df['item_price'].str.strip()
df['item_price'] = pd.to_numeric(df['item_price'], errors='coerce')

# Создаем фигуру с двумя графиками
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# График 1: Обычный scatter plot
scatter = ax1.scatter(df['quantity'], df['item_price'],
                     c=df['quantity'], cmap='viridis',
                     alpha=0.7, s=50, edgecolors='black', linewidth=0.3)
ax1.set_title('Scatter plot: Стоимость vs Количество', fontsize=14, fontweight='bold')
ax1.set_xlabel('Количество', fontsize=12)
ax1.set_ylabel('Стоимость товара ($)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
plt.colorbar(scatter, ax=ax1, label='Количество')

# График 2: Box plot по группам количества
quantity_groups = []
prices_by_quantity = []

for qty in sorted(df['quantity'].unique()):
    prices = df[df['quantity'] == qty]['item_price']
    quantity_groups.append(qty)
    prices_by_quantity.append(prices)

box_plot = ax2.boxplot(prices_by_quantity, labels=quantity_groups, patch_artist=True)

# Настройка цветов box plot
colors = plt.cm.Set3(np.linspace(0, 1, len(quantity_groups)))
for patch, color in zip(box_plot['boxes'], colors):
    patch.set_facecolor(color)

ax2.set_title('Box plot: Распределение стоимости по количеству', fontsize=14, fontweight='bold')
ax2.set_xlabel('Количество', fontsize=12)
ax2.set_ylabel('Стоимость товара ($)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# Детальный анализ корреляции
correlation = df['quantity'].corr(df['item_price'])
print(f"\nКоэффициент корреляции между количеством и стоимостью: {correlation:.3f}")

if correlation > 0.7:
    print("Сильная положительная корреляция")
elif correlation > 0.3:
    print("Умеренная положительная корреляция")
elif correlation > -0.3:
    print("Слабая корреляция")
elif correlation > -0.7:
    print("Умеренная отрицательная корреляция")
else:
    print("Сильная отрицательная корреляция")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Чтение файла с данными
df = pd.read_csv('chipotle.tsv', sep='\t', on_bad_lines='skip')

# Очистка столбца 'item_price' от символов '$', пробелов и запятых
df['item_price'] = df['item_price'].astype(str)
df['item_price'] = df['item_price'].str.replace('$', '', regex=False)
df['item_price'] = df['item_price'].str.replace(',', '', regex=False)
df['item_price'] = df['item_price'].str.strip()
df['item_price'] = pd.to_numeric(df['item_price'], errors='coerce')

# График 1: Основной scatter plot
plt.figure(figsize=(12, 8))

# Создаем scatter plot с прозрачностью для лучшей визуализации перекрывающихся точек
plt.scatter(df['quantity'], df['item_price'],
           alpha=0.6, color='blue', s=30, edgecolors='black', linewidth=0.5)

# Настройки графика
plt.title('Зависимость стоимости товара от количества', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Количество', fontsize=12)
plt.ylabel('Стоимость товара ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Добавляем линию тренда
z = np.polyfit(df['quantity'], df['item_price'], 1)
p = np.poly1d(z)
plt.plot(df['quantity'], p(df['quantity']), "r--", alpha=0.8, linewidth=2,
         label=f'Линия тренда: y = {z[0]:.2f}x + {z[1]:.2f}')

plt.legend()

# Улучшаем внешний вид
plt.tight_layout()

# Сохраняем в PDF
plt.savefig('scatter_plot_quantity_vs_price.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()

# График 2: Улучшенная версия с группировкой
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# График 2.1: Scatter plot с цветовым кодированием
scatter = ax1.scatter(df['quantity'], df['item_price'],
                     c=df['quantity'], cmap='viridis',
                     alpha=0.7, s=50, edgecolors='black', linewidth=0.3)
ax1.set_title('Scatter plot: Стоимость vs Количество', fontsize=14, fontweight='bold')
ax1.set_xlabel('Количество', fontsize=12)
ax1.set_ylabel('Стоимость товара ($)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
plt.colorbar(scatter, ax=ax1, label='Количество')

# График 2.2: Box plot по группам количества
quantity_groups = []
prices_by_quantity = []

for qty in sorted(df['quantity'].unique()):
    prices = df[df['quantity'] == qty]['item_price']
    quantity_groups.append(qty)
    prices_by_quantity.append(prices)

box_plot = ax2.boxplot(prices_by_quantity, labels=quantity_groups, patch_artist=True)

# Настройка цветов box plot
colors = plt.cm.Set3(np.linspace(0, 1, len(quantity_groups)))
for patch, color in zip(box_plot['boxes'], colors):
    patch.set_facecolor(color)

ax2.set_title('Box plot: Распределение стоимости по количеству', fontsize=14, fontweight='bold')
ax2.set_xlabel('Количество', fontsize=12)
ax2.set_ylabel('Стоимость товара ($)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()

# Сохраняем комбинированный график в PDF
plt.savefig('combined_plots_quantity_price.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()

# График 3: Scatter plot с группировкой по товарам
plt.figure(figsize=(14, 8))

# Берем топ-5 самых популярных товаров для цветового кодирования
top_items = df['item_name'].value_counts().head(5).index
df['item_category'] = df['item_name'].apply(lambda x: x if x in top_items else 'Другие товары')

# Создаем цветовую схему для категорий товаров
colors = {'Chicken Bowl': 'red', 'Burrito': 'blue', 'Chips and Guacamole': 'green',
          'Steak Burrito': 'orange', 'Canned Soft Drink': 'purple', 'Другие товары': 'gray'}

for category, color in colors.items():
    category_data = df[df['item_category'] == category]
    plt.scatter(category_data['quantity'], category_data['item_price'],
               c=color, label=category, alpha=0.7, s=40, edgecolors='black', linewidth=0.3)

plt.title('Зависимость стоимости от количества (с группировкой по товарам)',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Количество', fontsize=12)
plt.ylabel('Стоимость товара ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

# Сохраняем график с группировкой в PDF
plt.savefig('scatter_plot_by_items.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()

# Дополнительная статистика
print("=" * 60)
print("СТАТИСТИКА ПО ЗАВИСИМОСТИ СТОИМОСТИ ОТ КОЛИЧЕСТВА")
print("=" * 60)
print(f"Общее количество наблюдений: {len(df)}")
print(f"Уникальные значения количества: {sorted(df['quantity'].unique())}")
print(f"Максимальное количество: {df['quantity'].max()}")
print(f"Минимальное количество: {df['quantity'].min()}")
print(f"Средняя стоимость товара: ${df['item_price'].mean():.2f}")
print(f"Медианная стоимость товара: ${df['item_price'].median():.2f}")

# Анализ корреляции
correlation = df['quantity'].corr(df['item_price'])
print(f"\nКоэффициент корреляции между количеством и стоимостью: {correlation:.3f}")

if correlation > 0.7:
    print("Сильная положительная корреляция")
elif correlation > 0.3:
    print("Умеренная положительная корреляция")
elif correlation > -0.3:
    print("Слабая корреляция")
elif correlation > -0.7:
    print("Умеренная отрицательная корреляция")
else:
    print("Сильная отрицательная корреляция")

print("\nГрафики сохранены в файлы:")
print("1. scatter_plot_quantity_vs_price.pdf - основной scatter plot")
print("2. combined_plots_quantity_price.pdf - комбинированный график")
print("3. scatter_plot_by_items.pdf - scatter plot с группировкой по товарам")
```
